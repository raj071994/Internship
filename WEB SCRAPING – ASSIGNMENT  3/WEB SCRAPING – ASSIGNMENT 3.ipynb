```python
#fetch url for 3 pages if present
urls=[]
start=0
end=3

for page in range(start,end):
    time.sleep(4)
    for i in driver.find_elements_by_xpath('//a[@class="a-link-normal s-link-style a-text-normal"]'):
        urls.append(i.get_attribute("href"))
    nxt_button = driver.find_elements_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    try:
        driver.get(nxt_button[1].get_attribute('href'))
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
```

```python
len(urls)
```

```python
#define variable
brand_name =[]
product_name=[]
price_detail=[]
return_exchange=[]
expected_delivery=[]
availibility=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    #fectch brand name
    
    try:
        brand=driver.find_element_by_xpath('//span[@id="productTitle"]')
        brand_name.append(brand.text.split(" ")[0])
    except:
        brand_name.append('-')
        
    #fectch product name
    try:
        product=driver.find_element_by_xpath('//span[@id="productTitle"]')
        product_name.append(product.text)
    except:
        product_name.append('-')
        
    #fectch price details
    try:
        price=driver.find_element_by_xpath('//span[@id="tp_price_block_total_price_ww"]/span')
        price_detail.append(price.get_attribute('innerHTML'))
    except:
        price_detail.append('-')
        
     #fectch return/exchange
    try:
        rerurn=driver.find_element_by_xpath('//div[@class="a-section a-spacing-none icon-content"]/a')
        return_exchange.append(rerurn.text)
    except:
        return_exchange.append('-')
        
    #fectch expected delivery time
    try:
        delivery=driver.find_element_by_xpath('//div[@class="a-section a-spacing-mini"]/b')
        expected_delivery.append(delivery.text)
    except:
        expected_delivery.append('-')
        
    #fectch Availibility
    try:
        available=  driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        availibility.append(available.text)
    except:
        availibility.append('-')
  
        
```

```python
print(len(urls))
print(len(brand_name))
print(len(product_name))
print(len(price_detail))
print(len(return_exchange))
print(len(expected_delivery))
print(len(availibility))
```

```python
products = pd.DataFrame()
products['Brand Name']= brand_name
products['Product Nam']= product_name
products['Price']= price_detail
products['Return/Exchange']= return_exchange
products['Expected Delivery']= expected_delivery
products['Availibility']= availibility
products['URL']= urls
products
```

```python
Search_button = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
Search_button.click()
```

```python
#fetch url
urls=[]
for i in driver.find_elements_by_xpath('//a[@class="_1fQZEK"]'):
        urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
#define variable
brand_name =[]
product_name=[]
color_detail=[]
ram_details=[]
strorage_Rom_details=[]
primary_camera_details=[]
secondary_camera_details=[]
display_detail=[]
battery_cap=[]
price_details=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    #fectch brand name
    
    try:
        brand=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        brand_name.append(brand.text.split(" ")[0])
    except:
        brand_name.append('-')
        
    #fectch product name
    try:
        product=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        product_name.append(product.text)
    except:
        product_name.append('-')
                
    #fectch color details
    try:
        color=driver.find_elements_by_xpath('//li[@class="_21lJbe"]')
        for value in range(0,7):
            if(value == 3):
                color_detail.append(color[value].text)
    except:
        color_detail.append('-')
        
    #fectch ram/rom/camera details
    details=driver.find_elements_by_xpath('//li[@class="_21Ahn-"]')
    for value in range(0,len(details)):
        try:
            if(value == 0 and "|" in details[value].text):
                ram_details.append(details[value].text.split("|")[0])
        except:
            ram_details.append('-')
        
        #fectch rom details
        try:
            if(value == 0 and "|" in details[value].text):
                strorage_Rom_details.append(details[value].text.split("|")[1])
        except:
            strorage_Rom_details.append('-')
        
         #fectch front camera details
        try:
            if(value == 2 and "|" in details[value].text):
                primary_camera_details.append(details[value].text.split("|")[1])
            elif(value == 2):
                primary_camera_details.append('-')
        except:
            primary_camera_details.append('-')
            
        #fectch back camera details
        try:
            if(value == 2 and "|" in details[value].text):
                secondary_camera_details.append(details[value].text.split("|")[0])
            elif (value == 2):
                secondary_camera_details.append(details[value].text)
        except:
            secondary_camera_details.append('-')
            
        #fectch display details
        try:
            if(value == 1):
                display_detail.append(details[value].text)
        except:
            display_detail.append('-')
            
            
        #fectch front camera details
        try:
            if(value == 3):
                battery_cap.append(details[value].text)
            elif(len(details)<=3 and value == 2):
                battery_cap.append(details[value].text)
        except:
            battery_cap.append('-')
               
     #fectch price details
    try:
        price=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price_details.append(price.get_attribute('innerHTML'))
    except:
        price_details.append('-')
        
    
```

```python
print(len(urls))
print(len(brand_name))
print(len(product_name))
print(len(color_detail))
print(len(ram_details))
print(len(strorage_Rom_details))
print(len(primary_camera_details))
print(len(secondary_camera_details))
print(len(display_detail))
print(len(battery_cap))
print(len(price_details))
```

```python
smartphone = pd.DataFrame()
smartphone['Brand Name']= brand_name
smartphone['Product Nam']= product_name
smartphone['Color']= color_detail
smartphone['Ram']= ram_details
smartphone['Storage Rom']= strorage_Rom_details
smartphone['Primary Camera']= primary_camera_details
smartphone['Secondary Camera']= secondary_camera_details
smartphone['Display']= display_detail
smartphone['Battry Capacity']= battery_cap
smartphone['Price']= price_details
smartphone['URL']= urls
smartphone
```

```python
smartphone.to_csv('smartphone.csv', index=False)
```
